```python

#Pide al usuario una temperatura. 
#Luego, pregúntale en qué unidades está la temperatura, Celsius o Fahrenheit. 
#Tu programa debe convertir la temperatura a la otra unidad. 
#Las conversiones son F = 9/5C + 32 y C = 5/9(F − 32).

print("Ingrese temperatura")
T = float(input("Temperatura: " ))
unidades = input("tu temperatura esta en Celsius o Fahrenheit: " )
if unidades.lower() == "celsius":
   T_F = ((9 / 5) * T ) + (32) 
   print("Temperatura en Fahrenheit es:", T_F)
elif unidades.lower() == "fahrenheit":
   T_C = ((5/ 9) *( T  - (32)) )
   print("Temperatura en Celsius es:", T_C)
else:
   print("No se reconocio ninguna unidad")
   
```

```python

#A simple way of encrypting a message is to rearrange its characters. One way to rearrange the
#characters is to pick out the characters at even indices, put them first in the encrypted string,
#and follow them by the odd characters. For example, the string message would be encrypted
#as msaeesg because the even characters are m, s, a, e (at indices 0, 2, 4, and 6) and the odd
#characters are e, s, g (at indices 1, 3, and 5).
#(a) Write a program that asks the user for a string and uses this method to encrypt the string.
#(b) Write a program that decrypts a string that was encrypted with this method.

#A
cadena = input("Ingresa una cadena: ")
c = cadena
long = len(c)
pares = ""
impares = ""

for i in range(long):  # Corregimos el bucle y usamos range para iterar sobre los índices
    if i % 2 == 0:
        pares += c[i]
    else:
        impares += c[i]

print("Caracteres en posiciones pares:", pares + impares)

#B
cadena = input("Ingresa una cadena: ")
long = len(cadena)

primeramitad = ""  # Inicializa las cadenas vacías
segundamitad = ""

if long % 2 == 0:  # Verifica si la longitud es par
    mitad = long // 2
    primeramitad = cadena[:mitad]
    segundamitad = cadena[mitad:]
else:  # Si la longitud es impar
    mitad = (long - 1) // 2
    primeramitad = cadena[:mitad + 1]
    segundamitad = cadena[mitad + 1:]
print("La primera mitad de la cadena es:", primeramitad)
print("La segunda mitad de la cadena es:", segundamitad)

long1 = len(primeramitad)
long2 = len(segundamitad)
resultado = ""

for i in range(min(long1, long2)):
    resultado += primeramitad[i] + segundamitad[i]
if len(primeramitad) % 2 == 0 and len(segundamitad) % 2 == 0:  #Verifica si la longitud es par
    print(resultado)
else:
    print(resultado + primeramitad[-1] )
```
