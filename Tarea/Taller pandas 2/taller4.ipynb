```python


# Carga el archivo CSV en un DataFrame
df = pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia._20231031.csv')

# Itera a través de las columnas y unifica los valores
for columna in df.columns:
    if df[columna].dtype == 'object':  # Verifica si la columna es de tipo cadena
        df[columna] = df[columna].str.upper()  # Convierte a mayúsculas
        df[columna] = df[columna].str.replace('VALOR_A_UNIFICAR', 'VALOR_UNIFICADO')

df.head(3)
```

```python
df.isnull().sum()
columnas_a_reemplazar = ['Fecha de muerte', 'Código ISO del país','Nombre del grupo étnico'] # se escogieron estas 3 columnas pues luego de analizar, eran las 3 que contenian mas valores nulos
df[columnas_a_reemplazar] = df[columnas_a_reemplazar].fillna(0)
df['Nombre del país'].fillna("Colombia", inplace=True)  #se agrego el nombre del pais, solo para probar 
df.head(3)
```

```python

# Carga el archivo CSV en un DataFrame
df = pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia._20231031.csv')

# Lista de las columnas de fecha a procesar
columnas_de_fecha = ['Fecha de notificación', 'Fecha de inicio de síntomas', 'Fecha de muerte', 'Fecha de diagnóstico', 'Fecha de recuperación']

# Itera a través de las columnas de fecha y realiza la conversión y creación de nuevas columnas
for columna in columnas_de_fecha:
    df[columna] = pd.to_datetime(df[columna])  # Convierte la columna al formato datetime
    df[f'Día {columna}'] = df[columna].dt.day  # Crea una columna con el día
    df[f'Mes {columna}'] = df[columna].dt.month  # Crea una columna con el mes
    df[f'Año {columna}'] = df[columna].dt.year  # Crea una columna con el año

df.head(3)
#pueden aparecer datos nulos se debe a que no hay fechas ahi
```

```python

# Carga el archivo CSV en un DataFrame
df = pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia._20231031.csv')

# Filtra las filas del departamento "VALLE"
departamento_elegido = "VALLE"
df_valle = df[df['Nombre departamento'] == departamento_elegido]

# Obtiene la lista de ciudades únicas en el departamento VALLE
ciudades = df_valle['Nombre municipio'].unique()

# Crea vectores de NumPy para almacenar los datos de fallecidos, graves, leves y moderados
fallecidos_por_ciudad = np.zeros(len(ciudades), dtype=int)
graves_por_ciudad = np.zeros(len(ciudades), dtype=int)
leves_por_ciudad = np.zeros(len(ciudades), dtype=int)
moderados_por_ciudad = np.zeros(len(ciudades), dtype=int)

# Itera a través de las filas del DataFrame del departamento VALLE
for index, row in df_valle.iterrows():
    ciudad = row['Nombre municipio']
    indice_ciudad = np.where(ciudades == ciudad)[0][0]

    if row['Estado'] == 'Fallecido':
        fallecidos_por_ciudad[indice_ciudad] += 1
    elif row['Estado'] == 'Grave':
        graves_por_ciudad[indice_ciudad] += 1
    elif row['Estado'] == 'Leve':
        leves_por_ciudad[indice_ciudad] += 1
    elif row['Estado'] == 'Moderado':
        moderados_por_ciudad[indice_ciudad] += 1
df.head(3)

# Ahora tienes vectores NumPy con la información deseada por ciudad en el departamento VALLE
```

```python

df = pd.read_csv('Casos_positivos_de_COVID-19_en_Colombia._20231031.csv')

# Convierte las columnas de fecha al formato datetime
df['Fecha de notificación'] = pd.to_datetime(df['Fecha de notificación'])

# Encuentra la primera fecha en el DataFrame
primera_fecha = df['Fecha de notificación'].min()

# Calcula los días transcurridos desde la primera fecha
df['Días transcurridos'] = (df['Fecha de notificación'] - primera_fecha).dt.days

# Filtra los datos para obtener muertes y recuperaciones de hombres y mujeres
muertes_hombres = df[(df['Sexo'] == 'M') & (df['Estado'] == 'Fallecido')]
muertes_mujeres = df[(df['Sexo'] == 'F') & (df['Estado'] == 'Fallecido')]
recuperaciones_hombres = df[(df['Sexo'] == 'M') & (df['Estado'] == 'Recuperado')]
recuperaciones_mujeres = df[(df['Sexo'] == 'F') & (df['Estado'] == 'Recuperado')]

# Agrupa y cuenta las muertes y recuperaciones por días transcurridos
muertes_hombres_por_dia = muertes_hombres.groupby('Días transcurridos').size().values
muertes_mujeres_por_dia = muertes_mujeres.groupby('Días transcurridos').size().values
recuperaciones_hombres_por_dia = recuperaciones_hombres.groupby('Días transcurridos').size().values
recuperaciones_mujeres_por_dia = recuperaciones_mujeres.groupby('Días transcurridos').size().values

# Llena los valores NaN con ceros en caso de que haya días sin datos
dias_max = df['Días transcurridos'].max() + 1
datos_dias = np.arange(dias_max)

muertes_hombres_por_dia = np.bincount(muertes_hombres['Días transcurridos'], minlength=dias_max)
muertes_mujeres_por_dia = np.bincount(muertes_mujeres['Días transcurridos'], minlength=dias_max)
recuperaciones_hombres_por_dia = np.bincount(recuperaciones_hombres['Días transcurridos'], minlength=dias_max)
recuperaciones_mujeres_por_dia = np.bincount(recuperaciones_mujeres['Días transcurridos'], minlength=dias_max)

# Crea la matriz de NumPy con la información
matriz_informacion = np.column_stack((datos_dias, muertes_hombres_por_dia, muertes_mujeres_por_dia, recuperaciones_hombres_por_dia, recuperaciones_mujeres_por_dia))
print(matriz_informacion)

# La matriz contiene: [Días, Muertes Hombres, Muertes Mujeres, Recuperaciones Hombres, Recuperaciones Mujeres]



```

```python
# Asegurarse de que la columna de fechas esté en el formato datetime
df['fecha reporte web'] = pd.to_datetime(df['fecha reporte web'])

# Encontrar la fecha más temprana en la columna
fecha_inicial = df['fecha reporte web'].min()

# Calcular la diferencia de días entre cada fecha y la fecha más temprana
dias_transcurridos = (df['fecha reporte web'] - fecha_inicial).dt.days

# Almacenar los valores en una matriz de NumPy
matriz_dias_transcurridos = dias_transcurridos.to_numpy()
print(matriz_dias_transcurridos)

# 'matriz_dias_transcurridos' ahora contiene los días transcurridos desde la primera fecha
```

```python
# Asegurarse de que la columna de fechas esté en el formato datetime
df['fecha reporte web'] = pd.to_datetime(df['fecha reporte web'])

# Calcular la columna de días transcurridos (como se mencionó anteriormente)

# Filtrar los registros de hombres que murieron
muertes_hombres = df[(df['Sexo'] == 'M') & (df['Estado'] == 'Fallecido')]

# Contar el número de muertes de hombres para cada fecha de la columna de días transcurridos
muertes_hombres_por_dia = muertes_hombres['fecha reporte web'].value_counts().sort_index()

matriz_muertes_hombres = muertes_hombres_por_dia.to_numpy()
print(matriz_muertes_hombres)
```
